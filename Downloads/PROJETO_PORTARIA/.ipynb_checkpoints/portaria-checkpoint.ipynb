```python
#IMPORTANTO AS BIBLIOTECAS APOS A INSTALAÇÃO

import sqlite3 as sql3
import cv2
import os
import time
import pyttsx3

from deepface import DeepFace

import PIL
from PIL import Image
```

```python
# CRIANDO O CAMINHO DAS PASTAS DAS IMAGENS
foto_path = os.path.join(os.getcwd(), 'fotos')
foto_db = os.path.join(os.getcwd(), 'db_fotos')
```

```python
# CRIANDO AS FUNÇÕES :
# Falar
def falar(texto):  
    maquina = pyttsx3.init()
    rate = maquina.getProperty('rate')
    maquina.setProperty('rate', 200)
    maquina.say(texto)
    maquina.runAndWait()
    maquina.stop()
#---------------------------------------------------------------------------    
# Capturar Imagem
def capturarImagem(cpf):
    # Define o tamanho desejado para a janela
    w = 960
    h = 700
    winName = 'App Visitantes'
    cv2.namedWindow(winName, cv2.WINDOW_AUTOSIZE)
    # Posiciona a janela na metade direita do (meu) monitor
    cv2.moveWindow(winName, 100, 100)

    WebCamera = cv2.VideoCapture(0)
    amostra = 0
    numeroAmostras = 1    
    falar('Favor Olhar para a Camera')
    #foto='foto'
    while True:
        conectado, imagem = WebCamera.read()
        
        if cv2.waitKey(1) & 0xFF == ord('w'):
            cv2.imwrite("fotos/" + cpf + ".jpg", imagem) 
            falar('Foto capturada com sucesso')
            print("[Foto" + " capturada com sucesso]")
            amostra += 1            
        # Altera o tamanho da imagem para o desejado
        imagem = cv2.resize(imagem, (w, h))
        cv2.imshow(winName, imagem)
        if (cv2.waitKey(1) == ord('q') or amostra >= numeroAmostras):           
            break
    WebCamera.release()
    cv2.destroyAllWindows()
    foto= cpf + ".jpg"
    verificarVisitante(foto,cpf)    

#---------------------------------------------------------------------------
def verificarVisitante(fts,cpf):
    #print("01fts",fts)
    #print("02cpf",cpf)
    
    resultado= False
    print(resultado)
    try:       
        foto= os.path.join(foto_path,fts)        
        foto_db=selectCpf(cpf)
        result = DeepFace.verify(img1_path = foto, img2_path = foto_db ,enforce_detection=False) 
        print(result['verified'])
        resultado= result['verified']
        print(resultado)         
        
        if(resultado == True):
            falar('Visitante detectado,Acesso Liberado')
            time.sleep(5)
            falar('Foto do Visitante no Banco de Dados')
            image = Image.open(foto_db)
            image.show()
            time.sleep(3) 
            
            falar('Foto do Visitante')
            image = Image.open(foto)
            image.show()
            time.sleep(3) 
            
                
            # Remover Fotos
            if os.path.isfile(foto):
                os.remove(foto) 
                time.sleep(3)
           # Remover Fotos
            if os.path.isfile(foto_db):
                os.remove(foto_db) 
                time.sleep(3)             
            
        else:            
                
            falar('ALERTA !!!Visitante Não foi detectado,favor fazer o Cadastro')
            falar('Foto no Banco de Dados')
            image = Image.open(foto_db)
            image.show()
            time.sleep(3) 
            falar('ALERTA !!!Visitante Não foi detectado,favor fazer o Cadastro')
            
            # Remover Fotos
            if os.path.isfile(foto_db):
                os.remove(foto_db) 
                time.sleep(3)  
            cpf = input("Digite o Cpf do Visitante:") 
            nome= input("Digite o Nome: ")            
            insertVisitante(int(cpf),nome,foto)
    except :
        falar('ERRO, Visitante Não foi detectado,Cadastro em andamento,Favor aguardadar um pouco') 
        #foto=os.path.join(foto_path,cpf)            
        cpf = input("Digite o Cpf do Visitante:") 
        nome= input("Digite o Nome: ")
        insertVisitante(int(cpf),nome,foto)
#--------------------------------------------

#-------------------------------------------
def convertToBinaryData(filename):
    # Convert digital data to binary format
    with open(filename, 'rb') as file:
        blobData = file.read()
    return blobData
#-------------------------------------------
def writeTofile(data, filename):
    # Convert binary data to proper format and write it on Hard Disk
    with open(filename, 'wb') as file:
        file.write(data)
    print("Imagem Capturada com sucesso: ", filename, "\n")
    #INSERINDO O VISITANTE NO BANCO DE IMAGENS
#--------------------------------------------
#Salvar o visitante no Banco de Imagens
def insertVisitante(cpf,name, photo):
    try:
        s='##'
        con = sql3.connect('cvt.db') 
        cursor = con.cursor()
        falar('Conectando a Base de Dados,Favor aguardar um pouco')
        print("Connected to SQLite")
        for i in range(5):
             s +=s
             print(s, end="")
             time.sleep(1)               
        
        sql_insert = 'insert into alunos(cpf,nome,foto)values (?, ?, ?)'        
        empPhoto = convertToBinaryData(photo)        
        data_tuple = (cpf,name, empPhoto)
        cursor.execute(sql_insert, data_tuple)
        con.commit()
        falar('Visitante Cadastrado com Sucesso,Acesso Liberado')
        print('')
        print("Visitante Cadastrado com Sucesso,Acesso Liberado")
        foto= os.path.join(foto_path,photo)
        image = Image.open(foto)
        image.show()        
        time.sleep(5)
        # Remover Fotos
        if os.path.isfile(foto):
            os.remove(foto) 
            time.sleep(3)        

        cursor.close()

    except sql3.Error as error:
        print("ERRO!! Ao Cadastrar", error)
    
#---------------------------------------------------------
def selectCpf(dados):
    try:
        s="##"
        con = sql3.connect('cvt.db') 
        cursor = con.cursor()
        falar('Conectando a Base de Dados,Favor aguardar um pouco')
        print("Connected to SQLite")
        for i in range(5):
             s +=s
             print(s, end="")
             time.sleep(1)

        sql_select = 'select * from alunos where cpf=?'
        # Dados converter
        selDados = [dados] 
        cursor.execute(sql_select, selDados)
        selectDados = cursor.fetchall()
        print()
        for row in selectDados:
            print("Id = ", row[0],"Cpf = ", row[1], "Name = ", row[2])
            cpf = row[1]
            name = row[2]
            photo = row[3]
            img= str(cpf) + ".jpg"
            photoPath =os.path.join(foto_db, img) 
            writeTofile(photo, photoPath)
            return photoPath
        cursor.close()

    except sql3.Error as error:
        print("ERRO!! no sqlite table", error)
        
#--------------------------------------------------------------------
   
```
